```
#Importing required packages
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from pnslib import utils
```

```
#Loading a coloured image
img = cv.imread('test.jpg')
```

```
#Displaying the coloured image
plt.imshow(img)
plt.xticks([]),plt.yticks([])
plt.show()
```

```
#Converting the image into grayscale image
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```
#Output of the grayscaled image
plt.imshow(gray)
plt.xticks([]),plt.yticks([])
plt.show()
```

```
#Normalization of the Grayscaled image
norm_img = np.zeros((300, 300))
norm_img = cv.normalize(gray, norm_img, 0, 255, cv.NORM_MINMAX)

plt.imshow(norm_img)
plt.xticks([]),plt.yticks([])
plt.show()
```

```
#face detection
face_cascade = cv.CascadeClassifier(utils.get_haarcascade_path('haarcascade_frontalface_default.xml'))
faces = face_cascade.detectMultiScale(norm_img, 1.3, 5)

for (x, y, w, h) in faces:
  cv.rectangle(norm_img, (x, y), (x+w, y+h), (255, 0, 0), 2);
```

```
#Output
plt.imshow(norm_img)
plt.xticks([]),plt.yticks([])
plt.show()
```
